```python
# agrupando grupo de categoria 
produto_categoria = vendas.groupby(["Product Category", "Product Name"])["Total Revenue"].agg(
    ["sum", "mean", "min", "max", "count"])


produto_categoria.columns=["Preco total", "Preco medio","Preco min", "Preco max", "Unid vendida"]
display(produto_categoria)
```

```python
# 1. Primeiro entendo os dados 
print("\n Visualizando as primeiras linhas do DataFrame de vendas:")
display(vendas.head(3))  # Mostra 3 linhas para ver a estrutura

# 2. Análise BÁSICA de quantidades vendidas por categoria
print("\n QUANTIDADE VENDIDA POR CATEGORIA:")
qtd_categoria = vendas.groupby('Product Category')['Units Sold'].sum()
display(qtd_categoria.to_frame(name='Total Vendido'))

print(f"\nTOTAL GERAL: {qtd_categoria.sum()} unidades")
```

```python
# Garantir que 'vendas' é um DataFrame pandas
vendas = pd.DataFrame(vendas)

# Converter a coluna 'Data' para o tipo datetime
vendas['Date'] = pd.to_datetime(vendas['Date'])

# Criar a coluna 'Periodo' com formato 'Mês/Ano'
vendas['Periodo'] = vendas['Date'].dt.strftime('%B %Y')

# Mostrar as primeiras linhas para verificação
display(vendas.head())
```

```python
vendas = vendas.drop('Date', axis=1)

display(vendas.head())
```

```python
# 3. Análise BÁSICA de faturamento por categoria
print("\n💰 FATURAMENTO POR CATEGORIA:")
fat_categoria = vendas.groupby('Product Category')['Total Revenue'].sum()
display(fat_categoria.to_frame(name='Faturamento Total').round(2))

print(f"\nFATURAMENTO TOTAL: R$ {fat_categoria.sum():.2f}")


```

```python
# 4. Análise por período (se existir a coluna 'Periodo' no DataFrame vendas)
if 'Periodo' in vendas.columns:
    print("\n VENDAS POR PERÍODO:")
    vendas_periodo = vendas.groupby('Periodo')['Units Sold'].sum()
    display(vendas_periodo.to_frame(name='Total Vendido'))
else:
    print("\nℹ️ Não encontrei a coluna 'Periodo' no DataFrame vendas")
```

```python
# 5. Análise completa (categoria + mês)
print("\n VISÃO DETALHADA (Categoria + Mês):")
resumo_completo = vendas.groupby(['Product Category', 'Periodo']).agg({
    'Units Sold': 'sum',
    'Total Revenue': 'sum'
}).rename(columns={'Units Sold': 'Total Vendido', 'Total Revenue': 'Faturamento'})
display(resumo_completo)
```

```python
# Agrupar vendas por região
analise_regiao = vendas.groupby(["Region", "Periodo"])[["Units Sold", "Total Revenue"]].sum()

# Visualizar vendas por região
print("Vendas nos Períodos por Região:")

display(analise_regiao)

# Somar vendas por região
print("\nTotal das vendas nos Períodos por Região:")
total_regiao = analise_regiao.groupby(["Region"]).sum()

# Visualizar total das vendas por região
display(total_regiao)

print(f"Total Geral das Vendas nos Períodos por Região:\n{total_regiao.sum()}")

print(f"\nPercentual %:\n{((total_regiao / total_regiao.sum()) * 100).round(2)}")
```
